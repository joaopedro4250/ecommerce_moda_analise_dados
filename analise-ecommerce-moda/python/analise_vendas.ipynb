```python
import pandas as pd
import numpy as np

df_vendas = pd.read_csv("../dados/women_clothing_ecommerce_sales.csv")

display(df_vendas)
```

```python
display(df_vendas)

display(df_vendas.shape)

display(list(df_vendas.columns))

display(df_vendas.info())

display(df_vendas.describe())
```

```python
df_vendas["order_date"] = pd.to_datetime(df_vendas["order_date"], format="%Y/%m/%d %H:%M:%S")
display(df_vendas.info())

df_vendas["size"] = df_vendas["size"].fillna("One Size")

df_vendas["sku"] = df_vendas["sku"].replace("bobo", np.nan)

df_valid = df_vendas[df_vendas["sku"].notna()].copy()
#criei esse df valid para ser mais exato para analise e dashboards, tinha valores como "bobo" no  sku, deixei como NaN 


df_valid.info()


display(df_valid)
```

```python
faturamento_total = df_vendas["revenue"].sum()
display(faturamento_total)

#faturamento por dia
faturamento_dia = df_vendas.groupby(df_vendas["order_date"].dt.date)["revenue"].sum()
display(faturamento_dia)

#quais são os 5 produtos(sku) que mais geraram faturamento
faturamento_por_sku = df_vendas.dropna(subset=["sku"]).groupby("sku")["revenue"].sum()
top_cinco = faturamento_por_sku.sort_values(ascending=False)
top_cinco = top_cinco.head(5)
display(top_cinco)

#preço médio de cada sku
preço_medio_sku = df_vendas.groupby("sku")["unit_price"].mean()
display(preço_medio_sku)

#quais pedidos tem mais de 3 itens no total
pedidos_tres_itens = df_vendas.groupby("order_id", as_index=False)["quantity"].sum().query("quantity > 3")
display(pedidos_tres_itens)

#anomalia de preços
df_anomaly = df_vendas[df_vendas["revenue"] != df_vendas["unit_price"] * df_vendas["quantity"]].copy()
#fiz esse anomaly para ver se tem algum valor que seja anomalia deixei separado para ser usado para ver se tem anomalias
#df_vendas.drop(columns=["value_ok", "sku_valido"], inplace=True)

#renomear o nome da coluna sku valido
#df_valid.rename(columns={"sku_valido": "sku_valid"}, inplace=True)
#df_valid = df_valid.drop(columns=["value_ok"])

#qual foi o size mais vendido em quantidades
size_mais_vendidoqtd = df_valid.groupby("size")["quantity"].sum().idxmax()

#faturamento total por cor
fat_cor_total = df_vendas.groupby("color", as_index=False)["revenue"].sum().sort_values("revenue", ascending=False)
#percebi dados de cores como "A" "B" "C" "D" mas nao tenho detalhe se foi error ou o padrao de nome para tal cor... melhor opção e nao mexer
 
display(fat_cor_total)
```

```python
vendas_alto_valor = df_valid.loc[df_valid["revenue"] > 500]
display(vendas_alto_valor)

#filtro expecifico escolhi para cor e tamanho
filtro_dkb_2xl = df_valid.loc[(df_valid["color"] == "Dark Blue") &(df_valid["size"] == "2XL")].head(10)
display(filtro_dkb_2xl)

#testando .iloc
amostra = df_valid.iloc[:5, :5]
display(amostra)

#criando nova tabela e usando np.where para dizer se e normal ou alta o valor
df_valid["categoria_venda"] = np.where(df_valid["revenue"] >= 300, "Alta", "Normal")
display(df_valid)

#vendas expecificas 
vendas_pos_junho = df_valid.loc[df_valid["order_date"] >= "2022-07-01"]
display(vendas_pos_junho)

vendas_pos_junho.to_csv("vendas_pos_junho.csv",index=False)
```
